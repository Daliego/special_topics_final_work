```python
import python_calamine
import pandas as pd
import numpy as np
from pandas import DataFrame
from pyexcel_ods import save_data
from collections import OrderedDict
```

```python
acarau_dataset = pd.read_excel("./data/original_data_aracau.ods", engine="odf")
iguatu_dataset = pd.read_excel("./data/original_data_iguatu.xlsx", engine="calamine")
quixada_dataset = pd.read_excel("./data/original_data_quixada.xlsx", engine="calamine")
```

```python
colunas_em_comum = set(acarau_dataset.columns).intersection(iguatu_dataset.columns).intersection(quixada_dataset.columns)
len(colunas_em_comum)
```

```python
colunas = set(acarau_dataset.columns).union(iguatu_dataset.columns).union(quixada_dataset.columns)
```

```python
colunas_exclusivas = colunas.difference(colunas_em_comum)
```

```python
exclusivas_acarau = colunas_exclusivas.intersection(acarau_dataset.columns)
exclusivas_iguatu = colunas_exclusivas.intersection(iguatu_dataset.columns)
exclusivas_quixada = colunas_exclusivas.intersection(quixada_dataset.columns)
```

```python
acarau_filtrado = acarau_dataset[colunas_novo_dataset]
quixada_filtrado = quixada_dataset[colunas_novo_dataset]
iguatu_filtrado = iguatu_dataset[colunas_novo_dataset]
```

```python
novo_dataset = pd.concat([acarau_filtrado, quixada_filtrado, iguatu_filtrado])
```
